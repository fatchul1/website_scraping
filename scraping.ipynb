Pada pekerjaan ini dilakukan Scraping Website dari Web Shpee dengan pencarian Macbook. Hasil dari project ini berupa file csv yang dapat digunakan untuk berbagi proses Data Analytics.



```python
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import time
```

```python
opsi = webdriver.ChromeOptions()
opsi.add_argument('--headless')
servis = Service ('chromedriver.exe')
```

```python
driver = webdriver.Chrome(service=servis, options = opsi)
```

```python
shopee_link = "https://shopee.co.id/search?keyword=macbook"
driver.set_window_size(1300,800)
```

```python
driver.get(shopee_link)

rentang = 500
for i in range (1,7):
    akhir = rentang * i
    perintah = "window.scrollTo(0,"+str(akhir)+")"
    driver.execute_script(perintah)
    print ("loading ke -" + str(i))
    time.sleep(i)

time.sleep(5)
```

```python
content = driver.page_source
```

```python
driver.save_screenshoot("home.png")
driver.quit()
```

```python
data = BeautifulSoup(content,'html.parser')
#print(data.encode("utf-8"))

i = 1
base_url = 'https://shopee.co.id'
list_nama, list_harga, list_terjual, list_lokasi = [], [], [], []

for area in data.find_all('div', class_="col-xs-2-4 shopee-search-item-result__item"):
    print('proses data ke-'+str(i))
    nama = area.find('div', class_="ie3A+n bM+7UW Cve6sh")
    if nama != None:
        nama = nama.get_text()
    harga = area.find('span', class_='ZEgDH9')
    if harga !=None:
        harga = harga.get_text()
    terjual = area.find('div', class_="r6HknA uEPGHT")
    if terjual != None:
        terjual = terjual.get_text()
    lokasi = area.find('div', class_="zGGwiV")
    if lokasi != None:
        lokasi = lokasi.get_text()
    
    list_nama.append(nama)
    list_harga.append(harga)
    list_terjual.append(terjual)
    list_lokasi.append(lokasi)
    
    i+=1
    print('-----')
    
df = pd.DataFrame({'Nama' : list_nama, 'Harga' : list_harga, 'Terjual' : list_terjual, 'Lokasi' : list_lokasi})
df.to_csv('macbook.csv', index = False)
df
```
